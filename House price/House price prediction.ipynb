```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('D:\VsCOde\ML by python\\homeprices.csv')
df
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Price Predction')
plt.scatter(df.area,df.price,color = 'r', marker = '*')
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.predict(np.array([[3100]]))
```

```python
m = model.coef_
```

```python
b = model.intercept_
```

```python
y = m*3100 + b
y
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Price Predction')
plt.scatter(df.area,df.price,color = 'r', marker = '*')
plt.plot(df.area, model.predict(df[['area']]), color = 'g')
```
